# Insight & Kesimpulan
## Analisis dan Prediksi Waktu Pengantaran Makanan

Proyek ini bertujuan untuk menganalisis faktor-faktor yang memengaruhi waktu pengantaran
makanan serta membangun model prediksi yang andal untuk memperkirakan waktu pengantaran.

### Insight Utama
- Traffic level memiliki pengaruh signifikan terhadap delivery time.
- Kondisi cuaca buruk cenderung meningkatkan waktu pengantaran.
- Jarak pengantaran dan waktu persiapan merupakan faktor numerik paling dominan.

### Performa Model
- Linear Regression memberikan performa terbaik dengan MAE ≈ 5.9 menit dan R² ≈ 0.83.
- Model yang lebih kompleks seperti Decision Tree dan Random Forest tidak memberikan
  peningkatan performa yang signifikan.

### Kesimpulan
Model Linear Regression dipilih sebagai model final karena memberikan keseimbangan terbaik
antara akurasi, stabilitas, dan interpretabilitas.

### Rekomendasi
- Optimalkan alokasi kurir pada jam dengan traffic tinggi.
- Perhitungkan faktor cuaca dalam estimasi waktu pengantaran.
- Gunakan model prediksi sebagai baseline estimasi ETA.
