# Food Delivery Time Analysis
## Problem Understanding & Problem Statement

## 1. Business Context

Layanan food delivery menghadapi tantangan dalam menjaga waktu pengiriman tetap cepat dan konsisten.
Keterlambatan pengiriman dapat menurunkan kepuasan pelanggan dan berdampak pada reputasi serta pendapatan platform.

Oleh karena itu, penting untuk memahami faktor-faktor yang memengaruhi delivery time agar operasional dapat dioptimalkan.

## 2. Problem Statement

### Primary Problem Statement
Apa saja faktor utama yang memengaruhi delivery time pada layanan food delivery?

### Secondary Problem Statement
Bagaimana cara memprediksi delivery time secara akurat berdasarkan karakteristik pesanan dan kondisi operasional?

## 3. Project Objectives

### Sebagai Data Analyst
- Mengidentifikasi faktor yang berkontribusi terhadap keterlambatan pengiriman.
- Menganalisis pola delivery time berdasarkan jarak, kondisi lalu lintas, cuaca, dan waktu pemesanan.
- Memberikan insight yang actionable untuk peningkatan performa delivery.

### Sebagai Data Scientist
- Membangun model prediksi delivery_time.
- Membandingkan beberapa algoritma machine learning.
- Menghasilkan model dengan performa yang andal dan stabil

## 4. Target Variable & Candidate Features

### Target Variable
- delivery_time (dalam menit)

### Candidate Features (sementara)
- distance
- traffic_condition
- weather
- preparation_time
- order_time
- courier_experience


## 5. Success Metrics

### Analisis Data
- Insight mampu menjelaskan faktor utama keterlambatan
- Visualisasi mudah dipahami oleh stakeholder non-teknis

### Machine Learning Model
- MAE sebagai metrik utama
- RMSE sebagai metrik pendukung
- Model tidak mengalami overfitting

## 6. Output Tahap 1

- Problem statement yang jelas dan relevan
- Tujuan proyek yang terukur dan realistis
- Target variabel telah ditentukan
- Siap melanjutkan ke tahap data preprocessing
