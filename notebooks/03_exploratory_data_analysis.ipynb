# Exploratory Data Analysis (EDA)
## Food Delivery Time Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/processed/food_delivery_clean.csv")

df.head()
```

```python
df.info()
```

## Distribusi Target Variable (Delivery_Time_min)

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Delivery_Time_min"], bins=30, kde=True)
plt.title("Distribusi Delivery Time (menit)")
plt.xlabel("Delivery Time (menit)")
plt.ylabel("Frekuensi")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Delivery_Time_min"])
plt.title("Boxplot Delivery Time")
plt.show()
```

## Hubungan Fitur Numerik dengan Delivery Time

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="Distance_km", y="Delivery_Time_min", data=df)
plt.title("Distance vs Delivery Time")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="Preparation_Time_min", y="Delivery_Time_min", data=df)
plt.title("Preparation Time vs Delivery Time")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="Courier_Experience_yrs", y="Delivery_Time_min", data=df)
plt.title("Courier Experience vs Delivery Time")
plt.show()
```

## Pengaruh Fitur Kategorikal terhadap Delivery Time

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Traffic_Level", y="Delivery_Time_min", data=df)
plt.title("Delivery Time berdasarkan Traffic Level")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Weather", y="Delivery_Time_min", data=df)
plt.title("Delivery Time berdasarkan Weather")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Time_of_Day", y="Delivery_Time_min", data=df)
plt.title("Delivery Time berdasarkan Time of Day")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Vehicle_Type", y="Delivery_Time_min", data=df)
plt.title("Delivery Time berdasarkan Vehicle Type")
plt.show()
```

## Korelasi Antar Fitur Numerik

```python
plt.figure(figsize=(8,6))
corr = df.select_dtypes(include=["int64", "float64"]).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap Korelasi Fitur Numerik")
plt.show()
```
